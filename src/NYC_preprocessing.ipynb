#### How much time is apart from check-in to check-in?
get average_gap. <br> Assumption: user has stayed in the same time zone all along

```python
def getDateTime(utc_time:str):
    return dt.datetime.strptime(utc_time, "%a %b %d %H:%M:%S +0000 %Y")
    
```

```python
utc_test = "Tue Apr 03 18:02:24 +0000 2012"
print(" ".join(utc_test.split(" ")[:3] + utc_test.split(" ")[-1:]))
```

```python
def getDate(utc_time:str):
    splited = " ".join(utc_time.split(" ")[:3] + utc_time.split(" ")[-1:])
    return dt.datetime.strptime(splited, "%a %b %d %Y")
```

```python
df['date'] = df['utcTime'].apply(getDate)
```

```python
df['datetime'] = df['utcTime'].apply(getDateTime)
```

```python
avg_diff = {}
for user in df['userId']:
    avg_diff[user] = pd.DataFrame(df.loc[df['userId'] == user]['datetime'].diff()[1:])['datetime'].mean().seconds/60
```

```python
plt.bar(list(avg_diff.keys()), avg_diff.values())
plt.show()
```

```python
num_dates = {}
for user in df['userId']:
    num_dates[user] = len(pd.DataFrame(df.loc[df['userId'] == user])['date'].unique())
```

```python
import collections
cnt = collections.Counter(list(num_dates.values()))
```
